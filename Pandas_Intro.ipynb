```python
subset_2 = my_stat.query('V2 != 10 | V4 >= 1')
# аблюдения, у которых значения переменной V2  не равняются 10, или значения переменной V4 больше или равно 1
```

В переменной my_stat лежат данные с которыми вам необходимо проделать следующее дейтвие. В этих данных (my_stat) создайте две новые переменных:

V5 = V1 + V4

V6 = натуральный логарифм переменной V2

```python
my_stat['V5'] = my_stat['V1'] + my_stat['V4']
```

```python
my_stat = my_stat.assign(V6 = np.log(my_stat.V2))
```

```python
my_stat = my_stat.rename(columns={'V1': 'session_value', 'V2': 'group', 'V3': 'time', 'V4': 'n_users'})
```

```python
my_stat['session_value'] = my_stat['session_value'].fillna(value=0) 
# В переменной session_value замените все пропущенные значения на нули
```

```python
median = my_stat.query('n_users >= 0').n_users.median()
```

```python
my_stat
```

```python
my_stat["n_users"] = np.where(my_stat["n_users"] < 0, median, my_stat["n_users"]) 
# numpy.where(condition[, x, y])
# Return elements chosen from x or y depending on condition.
```

```python
mean_session_value_data = my_stat.groupby('group', as_index=False).agg({'session_value': 'mean'}).rename(columns={'session_value': 'mean_session_value'})
```
